```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#загружаем данные
data=pd.read_csv('ALL_MEDALISTS.csv', skiprows=4)
data.head()
```

```python
def Gold(x):
    """
    Returns the number of rows from 'Gold'
    
    x: Series.
    return: int.
    """
    return x[x=='Gold'].count()

def Silver(x):
    """
    Returns the number of rows from 'Silver'
    
    x: Series.
    return: int.
    """
    return x[x=='Silver'].count()

def Bronze(x):
    """
    Returns the number of rows from 'Bronze'
    
    x: Series.
    return: int.
    """
    
    return x[x=='Bronze'].count()
```

## Постройте таблицу в которой по годам всех олимпиад покажите топовых атлетов США(1 атлет на год) по общему количеству медалей? Включите дисциплину атлета

```python
#нужные данные для формирования таблицы
data_select=data[data.NOC=='USA'][['Discipline','Edition','Athlete','Medal']]
```

```python
#формируем таблицу так, как она должна выглядеть для всех участников
pretable=data_select.groupby(['Edition','Athlete']).agg({'Discipline':'unique',
                                                    'Medal':[Bronze,Silver,Gold,'count']})

pretable=pretable.rename(index=str,columns={'unique':'','count': 'Total' })
```

```python
#оставляет только атлетов с максимальным количесвом медалей
table=pretable.groupby('Edition',group_keys=False)
table=table.apply(lambda x: x[x.Medal.Total==x.Medal.Total.max()])
table
```
